```python
# Group By

import pandas as pd
```

```python
df = pd.read_csv("weather_by_cities_group.csv")
df
```

```python
df_grpby = df.groupby("city")
df_grpby
```

```python
for city , data in df_grpby:
    print(city)
    print("=" *60)
    #print("\n")
    print(data)
    print("=" *60)
```

```python
# max
df_grpby.max()
```

```python
# count
df_grpby.count()
```

```python
for event , data in df_grpby_event:
    print(event)
    print(data)
```

# Join 

![](https://miro.medium.com/v2/resize:fit:1100/format:webp/1*yb76Gk03pZsjVDp79n2yKA.jpeg)

```python
t1= {
    "Courses" : ["Python","C","C++"],
    "Fee" : [20000,25000,30000],
    "Duration": ["30days","40days","50days"]
}

df1 = pd.DataFrame(t1,index =["m1","m2","m3"])
df1
```

```python
t2= {
    "Courses" : ["Django","Power BI","Excel"],
    "Discount" : [2000,3000,4500]
}

df2 = pd.DataFrame(t2,index =["m2","m3","m4"])
df2
```

```python
# Inner join

df_com = df1.join(df2,lsuffix="_dataframe1",rsuffix="_dataframe2",how="inner")
df_com
```

```python
# left join

df_com = df1.join(df2,lsuffix="_dataframe1",rsuffix="_dataframe2",how="left")
df_com
```

```python
# right join

df_com = df1.join(df2,lsuffix="_dataframe1",rsuffix="_dataframe2",how="right")
df_com
```

```python
# outer join

df_com = df1.join(df2,lsuffix="_dataframe1",rsuffix="_dataframe2",how="outer")
df_com
```

# Merge

```python
df3 = pd.DataFrame({
    "city" : ["delhi","mumbai","pune"],
    "temp": [40,35,30]
})
df3
```

```python
df4 = pd.DataFrame({
    "city" : ["delhi","paris","new york"],
    "humidity": [50,55,60]
})
df4
```

```python
# inner
df_merge = pd.merge(df3,df4,on="city", how = "inner")
df_merge
```

```python
# left
df_merge = pd.merge(df3,df4,on="city", how = "left")
df_merge
```

```python
# right
df_merge = pd.merge(df3,df4,on="city", how = "right")
df_merge
```

```python
# right
df_merge = pd.merge(df3,df4,on="city", how = "outer",indicator=True)
df_merge
```
