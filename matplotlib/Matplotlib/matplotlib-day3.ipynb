```python
import pandas as pd
```

```python
# Time Series Plot
dates = pd.date_range("2025-09-06",freq='D',periods=365)
dates
```

```python
np.random.seed(42)
```

```python
values = np.cumsum(np.random.randn(365)) + 100
values
```

```python
trends = np.linspace(0,20,365)
trends
```

```python
time_series = values + trends

plt.plot(dates,time_series, linewidth= 1.5)
#plt.fill_between(dates,time_series, alpha= 0.3)
plt.xlabel('Date')
plt.ylabel("value")
plt.tight_layout()
plt.show()
```

```python
time_series = values + trends

plt.plot(dates,time_series, linewidth= 1.5)
plt.fill_between(dates,time_series, alpha= 0.3)
plt.xlabel('Date')
plt.ylabel("value")
plt.xticks(rotation=45)
plt.grid(True,alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Heatmap

data = np.random.rand(10,12)
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
years = [f'Year { i}'for i in range(1,11)]
```

```python
im = plt.imshow(data,aspect='auto')
plt.colorbar(im)
plt.xticks(range(12),months)
plt.yticks(range(10),years)
plt.xlabel("Months")
plt.xlabel("years")

for i in range(10):
    for j in range(12):
        text= plt.text(j,i,f'{data[i,j]:.2f}',ha="center",va="center",color="black")


plt.show()
```
