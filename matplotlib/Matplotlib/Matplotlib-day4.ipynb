```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# load csv
df = pd.read_csv("gas_prices.csv")
df.head()
```

```python
year = df['Year'].tolist()
```

```python
year
```

```python
plt.figure(figsize=(8,5))
plt.title("Gas Price over time(in USD)",fontdict={"fontweight":"bold"})
plt.plot(year,df.USA,'b.-',label="USA")
plt.plot(year,df.Canada,'r.-',label="Canada")
plt.plot(year,df.Mexico,'g.-',label="Mexico")
plt.plot(year,df.France,'y.-',label="France")
plt.xticks(df['Year'][::3].tolist())
plt.xlabel("Year")
plt.ylabel("US Dollor")
plt.legend()
plt.show()         
```
