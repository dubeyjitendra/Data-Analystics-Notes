```python
import seaborn as sns
```

```python
sns.set_style()
```

```python
# PairPlot
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# to check the name of dataset

datasets = sns.get_dataset_names()
print(datasets)
```

```python
# load the dataset
iris_dataset = sns.load_dataset("iris")
```

```python
iris_dataset
```

```python
sns.pairplot(data= iris_dataset,hue="species")
plt.show()
```

```python
# heat

tips = sns.load_dataset("tips")
```

```python
tips
```

```python
corrected = tips.select_dtypes(include=["float64","int64"]).corr()
corrected
```

```python
sns.heatmap(corrected,annot=True,cmap="coolwarm", center=0)
plt.title("Correction heat map")
plt.show()
```

```python
# box plot
box= sns.load_dataset("titanic")
sns.boxplot(data=box,x = "age",y= "class",orient="h")
plt.show()
```

```python
box
```

```python
# Scatterplot
p = sns.load_dataset("penguins")
p
```

```python
sns.scatterplot(data=p,x="flipper_length_mm",y="body_mass_g", hue="species")
plt.show()
```

# Count Plot

```python
# Count Plot
countplot= sns.load_dataset("titanic")
countplot
```

```python
# Count Plot
sns.countplot(data=countplot,x = "class",hue="survived",palette="muted")
plt.show()
```

```python
# Stacked Plot
sns.catplot(data=countplot,x = "class",hue="survived",kind="count",palette="deep")
plt.show()
```

```python
# Regression plot

tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(data=tips,x = "total_bill", y= "tip",scatter_kws={"alpha":0.5},line_kws={"color":"red"})
plt.show()
```

```python
# facet Grid
p1 = sns.load_dataset("penguins")
p1
```

```python
g = sns.FacetGrid(data=p1,col="species",row="island", hue="sex")
g.map(sns.scatterplot,"flipper_length_mm","body_mass_g")
g.add_legend()
plt.show()
```
